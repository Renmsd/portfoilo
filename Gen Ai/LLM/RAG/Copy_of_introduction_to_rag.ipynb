```python
# loading your documents


from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("harrypotter1.txt")
documents = loader.load()
```

```python
len(documents)
```

```python
# chunking strategy
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(documents)
```

```python
# setting up a vector store

embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(docs, embeddings)
print(db.index.ntotal)
```

```python
# retrieval with score

query = "What happened in chapter 1"
docs = db.similarity_search(query)
```

```python
def gen_reply(query, context):
    from langchain.chat_models import ChatOpenAI
    from langchain.schema import (
        HumanMessage
    )
    model_name = "gpt-3.5-turbo"
    chat = ChatOpenAI(model_name=model_name, temperature=0)
    question = f"""
    You are an expert on the topic of Harry Potter.
    You are provided with a query and a context, please give a relevant answer.
    Query: {query}
    Context: {context}
"""

    response = chat([HumanMessage(content=question)]).content


    return response

gen_reply("What happened in chapter 2", str([i.page_content for i in docs]))
```
