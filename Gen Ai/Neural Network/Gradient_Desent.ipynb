```python
X= Variable(2*torch.ones(1), requires_grad=True )
Y= Variable(3*torch.ones(1), requires_grad=True )
```

```python
def cost_function(a,b):
    return a**2 + b**2
```

```python
cost_function(5,8)
```

```python
X.item()
```

```python
initail_cost = cost_function(X,Y)
initail_cost.item()
```

```python
initail_cost = cost_function(X,Y).item()

print(f'Original X: {X.item()}')
print(f'Original Y: {Y.item()}')
print(f'intial Cost: {initail_cost}')
print('-'*30)

z = cost_function(X,Y)
z.backward()
print(f'X.grad: {X.grad.item()}')
print(f'Y.grad: {Y.grad.item()}')

step = 0.1 # lr
print('-'*30)
X.data = X.data - step*X.grad.data
Y.data = Y.data - step*Y.grad.data

updated_cost = cost_function(X,Y).item()
print(f'Updated X: {X.item()}')
print(f'Updated Y: {Y.item()}')
print(f'Updated Cost: {updated_cost}')

X.grad.data.zero_()
Y.grad.data.zero_()
```

```python
step = 0.1
num_iteration = 20
x_values, y_values, z_values = [], [], []

for i in range(num_iteration):
    cost = cost_function(x, y)

    x_values.append(x.item())
    y_values.append(y.item())
    z_values.append(cost.item())

    print(f'Iteration: {i+1}')
    print(f'x: {x.item()}')
    print(f'y: {y.item()}')
    print(f'Cost: {cost.item()}')

    cost.backward()

    print(f'x Gradient: {x.grad.item()}')
    print(f'y Gradient: {y.grad.item()}')

    x.data = x.data - step * x.grad.data
    y.data = y.data - step * y.grad.data

    x.grad.data.zero_()
    y.grad.data.zero_()
    print('-'*30)
```
