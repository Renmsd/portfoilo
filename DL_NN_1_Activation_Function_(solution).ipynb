# 1. Activation Functions

Linear:  $$g(z) = z$$

Unit step: $$g(z) = \begin{cases} 0 & z < 0, \\ 1 & z \ge 0 \end{cases}$$

logistic (sigmoid): $$g(z) = \frac{1}{1 + e^{-z}}$$

hyperbolic tangent (sigmoid): $$g(z) = tanh(z) = \frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$$


Piecewise linear: $$g(z) = \begin{cases}
 1 & \mbox{for } z \geq \frac{1}{2} \\
 z + \frac{1}{2} & \mbox{for } -\frac{1}{2} < z < \frac{1}{2} \\
 0 & \mbox{for } z \leq -\frac{1}{2}
 \end{cases}$$

```python
import numpy as np
def activation(z, ftype):

    implemented = ('linear', 'unit step', 'logistic', 'piecewise linear', 'tanh', 'relu')
    if ftype not in implemented:
        raise AttributeError('%s not in %s' % (kind, implemented))

    if ftype == 'unit step':
        return np.where(z >= 0.0, 1, 0)

    elif ftype == 'logistic':
        return 1.0 / (1.0 + np.exp(-z))

    elif ftype == 'tanh':
        e_p = np.exp(z)
        e_m = np.exp(-z)
        return (e_p - e_m) / (e_p + e_m)

    elif ftype == 'piecewise linear':
        if z >= 0.5:
            return 1
        elif z <= -0.5:
            return 0
        else:
            return z + 0.5
    elif ftype == 'relu':
        if z <=0:
            return 0
        else:
            return z
    else:
        return z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.005)
for a in ('linear', 'unit step', 'logistic', 'piecewise linear', 'tanh', 'relu'):
    y = [activation(z, ftype=a) for z in x]
    plt.plot(x, y)
    plt.title(a)
    plt.ylim([-4, 4])
    plt.xlabel('z')
    plt.ylabel('f(x)')
    plt.grid()
    plt.show()
```
